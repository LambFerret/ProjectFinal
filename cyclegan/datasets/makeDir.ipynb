```python
# testA/B 사진 복사
import os
import shutil

origin = 'C:/Users/chohe/ProjectFinal/cyclegan/datasets/spring2summer/testA/'
copy = 'C:/Users/chohe/ProjectFinal/cyclegan/datasets/spring2fall/testA'

# oPath 경로의 파일 읽기
file_list = os.listdir(origin)

# 복사할 파일들을 닮을 배열 객체
copy_files = []

# Files List Create - TEXT 파일을 제외한 파일들이 이동대상 파일
for file in file_list:
	print("File : " + file)
	copy_files.append(file)

# Move Files
for cofi in copy_files:
	print('Copy File : ' + origin + ' 에 있는 ' + cofi + ' ----- > ' + copy + '로 이동하였습니다.')
	shutil.copy2(origin + '\\' + cofi, copy + '\\' + cofi)
print("[info] Copy Completed.")
```
